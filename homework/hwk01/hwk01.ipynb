## Q3 1.1
You can find a dataset giving the cost (in 1976 US dollars), number of megawatts, and year of construction of a set of
nuclear power plants at http://lib.stat.cmu.edu/DASL/Datafiles/NuclearPlants.html.

(a) Are there outliers in this data?

(b) What is the mean cost of a power plant? What is the standard deviation?

(c) What is the mean cost per megawatt? What is the standard deviation?

(d) Plot a histogram of the cost per megawatt. Is it skewed? Why?

```python
power = pd.read_csv('./q3.csv', names=['Cost', 'Mwatts', 'Date'], sep=',', header=0, index_col=False)
```

```python
# a
plt.subplot(1, 3, 1)
plt.boxplot(power['Cost'])
plt.title('Box Plots: Cost')
plt.ylabel('Cost')
plt.subplots_adjust(wspace = 1)
plt.subplot(1, 3, 2)
plt.boxplot(power['Mwatts'])
plt.title('Box Plots: Mwatts')
plt.ylabel('Mwatts')
plt.subplot(1, 3, 3)
plt.boxplot(power['Date'])
plt.title('Box Plots: Date')
plt.ylabel('Date')
plt.show()
```

```python
# b
print("---mean---")
print(power.mean(axis=0))
print("---std---")
print(power.std(axis=0))
```

```python
# c
print("---mean cost per megawatts---")
power['costPerW'] = (power['Cost']/power['Mwatts'])
print(power.mean(axis=0))
print("---std cost per megawatts---")
print(power.std(axis=0))
print("---median cost per megawatts---")
print(power.median(axis=0))
```

```python
# d
plt.hist(power['costPerW'], bins=10, edgecolor='black')
plt.ylabel('Power Plants Count')
plt.xlabel('Cost per MW')
plt.axvline(0.569735, color='red', linestyle='dashed', linewidth=1)
plt.axvline(0.57957, color='green', linestyle='dashed', linewidth=1)
plt.legend(['Mean', 'Median'])
plt.show()
```

# Q4 1.12
You can find a dataset giving the sodium content and calorie content of three types of hot dog at http://lib.stat.cmu.edu/DASL/Datafiles/Hotdogs.html. The types are Beef, Poultry, and Meat (a rather disturbingly vague label). Use class conditional histograms to compare these three types of hot dog with respect to sodium content and calories.

```python
import pandas as pd
import matplotlib.pyplot as plt

food = pd.read_csv('./q4.csv', names=['Type', 'Calories', 'Sodium'], sep=',', header=0, index_col=False)

beef = food[food['Type'] == 'Beef']
poultry = food[food['Type'] == 'Poultry']
meat = food[food['Type'] == 'Meat']

# Calories
print("---Calories count in each food category---")
plt.hist([beef['Calories'], poultry['Calories'], meat['Calories']], bins=15, label=['beef', 'poultry', 'meat'])
plt.legend(loc='upper right')
plt.title('Calories count in each food category')
plt.xlabel('Calories')
plt.ylabel('Food Count')
plt.show()

# Sodium
print("---Sodium count in each food category---")
plt.hist([beef['Sodium'], poultry['Sodium'], meat['Sodium']], bins=15, label=['beef', 'poultry', 'meat'])
plt.legend(loc='upper right')
plt.title('Sodium count in each food category')
plt.xlabel('Sodium')
plt.ylabel('Food Count')
plt.show()
```

# Q5 1.13
You will find a dataset giving (among other things) the number of 3 or more syllable words in advertising copy appearing in magazines at http://lib.stat.cmu.edu/DASL/Datafiles/magadsdat.html. The magazines are grouped by theeducation level of their readers; the groups are 1, 2, and 3 (the variable is called GRP in the data).

(a) Use a box plot to compare the number of three or more syllable words for the ads in magazines in these three groups. What do you see?

(b) Use a box plot to compare the number of sentences appearing in the ads in magazines in these three groups. What do you see?

```python
import pandas as pd
import matplotlib.pyplot as plt

magazine = pd.read_csv('./q5.csv', names=['WDS', 'SEN', '3SYL', 'MAG', 'GROUP'], sep=',', header=0, index_col=False)

# a
group1 = magazine[magazine['GROUP'] == 1]
group2 = magazine[magazine['GROUP'] == 2]
group3 = magazine[magazine['GROUP'] == 3]

plt.boxplot([group1['3SYL'], group2['3SYL'], group3['3SYL']])
plt.xlabel('Group')
plt.ylabel('Count')
plt.title('Three or more syllable word for each magazines groups')
plt.show()
```

```python
# b
plt.boxplot([group1['SEN'], group2['SEN'], group3['SEN']])
plt.xlabel('Group')
plt.ylabel('Count')
plt.title('Number of sentences for each magazines groups')
plt.show()
```
